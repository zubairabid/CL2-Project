```python
url = '../datasets/news-indicnlp/newsarticles/CLEANED/anandabazar_articles_clean.json'
url2 = '../datasets/news-indicnlp/newsarticles/CLEANED/ebala_articles_cleaner.json'
# url = 'https://raw.githubusercontent.com/zubairabid/CL2-Project/master/datasets/news-indicnlp/newsarticles/CLEANED/anandabazar_articles.json'

file1 = open(url, 'r', encoding = 'utf-8')
contents = file1.read()
file2 = open(url2, 'r', encoding = 'utf-8')
contents2 = file2.read()

df = pd.DataFrame(json.loads(contents))
df2 = pd.DataFrame(json.loads(contents2))

# df.append(df2)

df
```

```python
def tokenize(string):
    rlist = []
    punctuation = [' ', '।', '\'', '\"', ',', '‘', '’', '—']
    word = ''
    for character in string:
        if character in punctuation:
            rlist.append(word)
            word = ''
        else:
            word += character
    
    return rlist
```

```python
tdict = {}
```

```python
translator = Translator()
for i in range(1564):
    key = 'art'+str(i)
    
    if key in df:
        print('-'*50 + '\nInside article ' + key)
        for word in tokenize(df[key]['body']):
            print(word)
            
            if word not in tdict:
                print('Looking it up')
                tdict[word] = translator.translate(word, src='bn', dest='en')
                time.sleep(1)
            
            print(tdict[word].text)
            print()
            
```
