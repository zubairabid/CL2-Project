```python
import nltk
import pandas as pd
import json
from nltk.corpus import sentiwordnet as swn
from googletrans import Translator
```

```python
url = '../datasets/news-indicnlp/newsarticles/CLEANED/anandabazar_articles_clean.json'
# url = 'https://raw.githubusercontent.com/zubairabid/CL2-Project/master/datasets/news-indicnlp/newsarticles/CLEANED/anandabazar_articles.json'

file1 = open(url, 'r', encoding = 'utf-8')
contents = file1.read()
df = pd.DataFrame(json.loads(contents))

df
```

```python
def tokenize(string):
    rlist = []
    punctuation = [' ', '।', '\'', '\"', ',', '‘', '’', '—']
    word = ''
    for character in string:
        if character in punctuation:
            rlist.append(word)
            word = ''
        else:
            word += character
    
    return rlist
```

```python
translator = Translator()
for word in tokenize(df['art0']['body']):
    print(word)
    print(translator.translate(word, src='bn', dest='en').text)
```
