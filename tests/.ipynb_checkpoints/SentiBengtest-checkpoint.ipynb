```python
import nltk
import pandas as pd
import json
from nltk.corpus import sentiwordnet as swn
from googletrans import Translator
import pickle
import time
```

```python
url = '../datasets/news-indicnlp/newsarticles/CLEANED/anandabazar_articles_clean.json'
url2 = '../datasets/news-indicnlp/newsarticles/CLEANED/ebala_articles_cleaner.json'
# url = 'https://raw.githubusercontent.com/zubairabid/CL2-Project/master/datasets/news-indicnlp/newsarticles/CLEANED/anandabazar_articles.json'

file1 = open(url, 'r', encoding = 'utf-8')
contents = file1.read()
file2 = open(url2, 'r', encoding = 'utf-8')
contents2 = file2.read()

df = pd.DataFrame(json.loads(contents))
df2 = pd.DataFrame(json.loads(contents2))

# df.append(df2)

df
```

```python
df2
```

```python
newnums = ['art'+str(i) for i in range(1600, 8362)]
df2.columns = newnums
df2
```

```python
df = df.join(df2)
```

```python
df
```

```python
def tokenize(string):
    rlist = []
    punctuation = [' ', '।', '\'', '\"', ',', '‘', '’', '—']
    word = ''
    for character in string:
        if character in punctuation:
            rlist.append(word)
            word = ''
        else:
            word += character
    
    return rlist
```

```python
tdict = {}
```

```python
translator = Translator()
for i in range(1564):
    key = 'art'+str(i)
    
    if key in df:
        print('-'*50 + '\nInside article ' + key)
        for word in tokenize(df[key]['body']):
            print(word)
            
            if word not in tdict:
                print('Looking it up')
                tdict[word] = translator.translate(word, src='bn', dest='en')
                time.sleep(1)
            
            print(tdict[word].text)
            print()
            
```

```python
with open('filename.pickle', 'wb') as handle:
    pickle.dump(tdict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
with open('../filename.pickle', 'rb') as handle:
    b = pickle.load(handle)
    
for key in b:
    print(key, b[key])
    poswords = list(swn.senti_synsets(b[key]))
    print(poswords)
    for value in poswords:
        print(value, value.obj_score())
#         print(swn.senti_synset(value))
        print()
    print('-'*50)
    time.sleep(0.01)
```
