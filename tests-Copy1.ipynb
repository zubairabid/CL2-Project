```python
def tokenize(string):
    rlist = []
    punctuation = [' ', '।', '\'', '\"', ',', '‘', '’', '—']
    word = ''
    for character in string:
        if character in punctuation:
            rlist.append(word)
            word = ''
        else:
            word += character
    
    return rlist
```

```python
tokenize(df['art0']['body'])
```
